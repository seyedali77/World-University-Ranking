# Correlation Heatmap Analysis

The correlation heatmap displays relationships between various ranking and performance metrics such as "World Rank," "National Rank," and factors like "Quality of Education," "Alumni Employment," "Research Output," etc. Correlations are measured on a scale from -1 to +1, where:

- **1** indicates a perfect positive correlation.
- **-1** indicates a perfect negative correlation.
- **0** indicates no correlation.

## Key Observations:

### 1. World Rank vs. Score:
- **World Rank** has a strong negative correlation of **-0.91** with **Score**, indicating that as a university's world rank improves (lower rank number), its score tends to increase.

### 2. National Rank vs. Score:
- **National Rank** has a negative correlation of **-0.29** with **Score**, showing that a better national rank correlates moderately with a higher score.

### 3. World Rank and Performance Metrics:
- **World Rank** has strong positive correlations with:
  - **Research Output (0.88)**, **Quality Publications (0.95)**, and **Influence (0.88)**: Universities that perform better in research, publish more high-quality work, and have strong influence tend to have better world rankings.
  - **Citations (0.83)**: Universities with a higher number of citations also tend to have better world rankings.
  - Moderate positive correlation with **Alumni Employment (0.44)**.
  - Weak or negative correlations with **Quality of Faculty (-0.05)** and **Quality of Education (-0.16)**, suggesting that these factors may not impact world rankings as much.

### 4. Quality of Education:
- **Quality of Education** shows negative correlation with most metrics except **Quality of Faculty (0.28)** and **Score (0.05)**, suggesting that good faculty members contribute to better educational quality, while other performance measures like research output and influence may not directly align with education quality.

### 5. Alumni Employment:
- **Alumni Employment** correlates positively with **Research Output (0.35)**, **Quality Publications (0.39)**, and **Citations (0.35)**, meaning institutions that produce high-quality research also have better alumni employment outcomes.
- Moderate negative correlation with **Score (-0.50)** suggests institutions with higher alumni employment may not always score highly overall.

### 6. Quality of Faculty:
- **Quality of Faculty** correlates positively with **Quality of Education (0.28)** and **Alumni Employment (0.03)**, but shows low or negative correlations with most other metrics. Faculty quality impacts education and employment but may not affect other indicators like research output and influence.

### 7. Research Output and Other Metrics:
- **Research Output** has strong positive correlations with:
  - **Quality Publications (0.93)**: More research leads to more publications.
  - **Influence (0.68)**: Universities with higher research outputs have strong influence.
  - **Citations (0.65)**: More research leads to more citations.
  - **Score (-0.82)**: Despite positive research output, there’s a negative correlation with score, likely due to a complex weighting system in the ranking metrics.

### 8. Quality Publications and Citations:
- **Quality Publications** have high positive correlations with **Citations (0.75)**, **Influence (0.82)**, and **Research Output (0.93)**. Universities that publish high-quality research are often cited more and have a greater influence on the academic community.

### 9. Score vs. Influence, Research Output, Citations:
- **Score** has strong negative correlations with **Influence (-0.82)**, **Research Output (-0.82)**, **Citations (-0.79)**, and **Quality Publications (-0.87)**. Despite strong performance in research, citations, and influence, these factors may not directly lead to higher overall scores, suggesting other metrics like faculty or education quality weigh heavily in determining the final score.

## Conclusion:
- Institutions with high research output, citations, and quality publications tend to perform well in world rankings but may not always
